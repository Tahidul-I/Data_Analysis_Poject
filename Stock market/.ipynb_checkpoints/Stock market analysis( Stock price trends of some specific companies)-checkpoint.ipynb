```python
import pandas as pd
import matplotlib.pyplot as plt
stock_data = pd.read_csv('Stock_Market_Data.csv')
stock_data.head()
```

```python
#Exploratory Analysis ( i.e size of data, data types)
stock_data.shape
```

```python
stock_data.dtypes
```

```python
#Converting Date column filed type from object to time series
stock_data.Date = pd.to_datetime(stock_data.Date,dayfirst=True)
stock_data.dtypes
```

```python
#Determining how much data we have
stock_data.Date.max()
```

```python
#**So we have data on 6 months**
```

```python
#Selecting specific companies for price trend analysis
specific_compnies = ['UNILEVERCL','BEXIMCO','RECKITTBEN','BATBC','SAIHAMCOT']
# Making a new dataframe for consisting of the selected companies
selected_company_data = stock_data[stock_data['Name'].isin(specific_compnies)]
selected_company_data.head(35)
```

```python
#creating line graphs using for loop for dynamic plotting
plt.figure(figsize=(12,6))
for company in specific_compnies:
    closing_price = selected_company_data[selected_company_data['Name']==company].Close
    date = selected_company_data[selected_company_data['Name']==company].Date
    plt.plot(date,closing_price,label=company)
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.grid()
    plt.legend()
    plt.title('Stock price trends')
    plt.xticks(rotation=45)
plt.show()
    
```

```python
# From the above price trend, it can be seen that the stock mnarket closing date price for everyday for RECKITTBEN is very 
#fluctuating. So lets do a rolling average for every week and visualize it
selcted_company = 'RECKITTBEN'
selected_data = selected_company_data[selected_company_data['Name']==selcted_company]
selected_data['Week_Rolling Average']=selected_data['Close'].rolling(window=7).mean()
closing_price = selected_data['Close']
rolling_avg = selected_data['Week_Rolling Average']
date = selected_data['Date']
plt.figure(figsize=(12,6))
plt.plot(date,rolling_avg,label="Weekly Rolling average",color='blue')
plt.plot(date,closing_price,label="Closing price",color='red')
plt.legend()
plt.grid()
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```
